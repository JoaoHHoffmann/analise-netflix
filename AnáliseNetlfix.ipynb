## Analisando dados Dataset da netflix

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/content/netflix_data.csv')
```

##Para conhecer um pouco melhor o DataFrame, irei mostrar suas colunas, e os 10 primeiros elementos de cada uma das colunas.

```python
df.head(10)
```

## Filtrando dados de filmes que foram lançados nos anos 90

```python
filmes = df[df['type'] == "Movie"] # Selecionando somente os do tipo Filme(Movie)
filmes_90 = filmes[filmes['release_year'] <= 1999] # Selecionando os filmes que lançaram de 1999 para baixo.
filmes_decrescente = filmes_90.sort_values(by='release_year', ascending=False)# Ordenando os filmes de forma decrescente
filmes_decrescente
```

## Relações dos filmes lançados nos anos 90 e suas durações

```python
media_duracao = filmes_90['duration'].mean() # Pegando a média de duração dos filmes
plt.hist(filmes_90['duration'],edgecolor = 'black', color = 'lightblue') # Preparando o gráfico histograma, e fazendo algumas alterações nele
plt.title('Duração dos filmes lançados nos anos 90') # Titulo do gráfico
plt.xlabel('Duração(Minutos)') # Colocando uma legenda para o eixo X
plt.ylabel('Quantidade de filmes') # Colocando uma legenda para o eixo Y
plt.legend([f'Média de duração: {media_duracao:.1f} min']) # Colocando uma legenda mostrando a média de duração dos filmes, tempo em MINUTOS
plt.show() # Mostrando o gráfico
```

## Analisando os dados para ver qual os três países que mais produziram filmes na década de 90.

```python
paises_max_prod = filmes_90['country'].value_counts().head(5) # Pegando os 5 países com mais produções dos anos 90, e mostrando a quantidade de filmes de cada um deles
paises_max_prod
```

```python
plt.figure(figsize=(8, 5)) # Criando a figura do gráfico
plt.bar(paises_max_prod.index, paises_max_prod.values, color=['blue', 'green', 'red', 'yellow', 'orange']) # Criando as barras e alterando as cores da barras
plt.title('Países que mais produziram filmes na década de 90') # Colocando o titulo
plt.xlabel('País') # Colocando os valores do eixo x
plt.ylabel('Quantidade de filmes') # Colocando os valores do eixo y
plt.xticks(rotation=25) # Rotacionando os valores do eixo X em 25º
for i, valor in enumerate(paises_max_prod.values): # Esse laço de repetição percorre cada um dos 5 países e suas respectivas quantidades de filmes
    plt.text(i, valor + 2, str(valor), ha='center', fontsize=12, fontweight='bold')
plt.show()
```
